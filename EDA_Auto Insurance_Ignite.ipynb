## Confusion Matrix


- True Positive (TP) : The number of correct predictions where the actual response was “Yes”.
- True Negative (TN) : The number of correct predictions where the actual response was “No”.
- False Positive (FP) - Error Type 1 : The number of incorrect predictions where the mode predicted “Yes” but the actual response was “No”.
- False Negative (FN) - Error Type 2 : The number of incorrect predictions where the model predicted “No” but the actual response was “Yes”.

Based on the confusion matrix above, it can be said that:

**TN = Can avoid customers who really won't renew their insurance**

**TP = Possible to maintain loyal customers and increase loyal customers**

**Consequence of FN = The model predicts no response, which leads to a missed revenue opportunity from renew insurance**

**Consequence of FP = Estimated revenue loss occurs when the model incorrectly predicts a positive renewal response, leading to an overestimation of potential earnings. This misjudgment can result in reduced overall revenue as resources are focused on customers who ultimately do not renew..**

## Evaluation Metrics
To maximize revenue by gaining more policyholders, we decided to focus on evaluation metrics such as **recall and accuracy**. The primary goal is to minimize false negatives, as correctly identifying potential policyholders who will renew is crucial for increasing customer retention and, consequently, revenue. By emphasizing recall, we ensure the model captures as many potential renewals as possible, while accuracy helps to verify the overall effectiveness of the model in making correct predictions.
>Recall would be appropriate to ensure that we capture as many potential "yes" responses as possible, even at the expense of having more false positives. This might be the case if missing out on potential renewals is highly costly or if it is important to maximize customer retention.
>

>Accuracy would be appropriate to ensure a balanced performance in predicting both "yes" and "no" responses, providing a comprehensive view of the model's effectiveness. This metric is particularly useful when it’s important to maintain a high overall prediction quality, ensuring that both potential renewals and non-renewals are correctly identified to optimize customer retention and revenue growth.

#### Distribution of Numerical Data With Kolmogorov Smirnov

```python
from scipy import stats
# Selecting numeric columns from 'clv_model'
numeric_columns_clv_model = clv_model.select_dtypes(include=['number']).columns

# Conducting the Kolmogorov-Smirnov Test for each numeric column in 'clv_model'
test_results_clv_model = []
for column in numeric_columns_clv_model:
    stat, p_value = stats.kstest(clv_model[column], 'norm')
    conclusion = 'Not normally distributed' if p_value < 0.05 else 'Normally distributed'
    test_results_clv_model.append([column, stat, p_value, conclusion])

# Creating a DataFrame of the results
df_results_clv_model = pd.DataFrame(test_results_clv_model, columns=['Column', 'Statistic', 'P-Value', 'Conclusion'])

df_results_clv_model
```

### **Data Correlation**

##### **Pearson Correlation**

After analyzing the data distributions, we applied Pearson correlation to examine the relationships between our numeric features—such as 'Customer Lifetime Value,' 'Income,' and 'Total Claim Amount'—and the binary 'Response' variable, which indicates policy renewal. Pearson correlation is useful for quantifying the strength and direction of the linear relationship between a continuous variable and a binary outcome. This method allowed us to assess how changes in continuous variables are associated with the likelihood of renewal, providing actionable insights for predicting customer behavior and optimizing strategies for customer retention ([Penn State Eberly College of Science](https://online.stat.psu.edu/stat501/lesson/4/4.1), [UCLA IDRE](https://stats.idre.ucla.edu/spss/output/correlation/)).


```python
# Make a copy of the dataframe 'clv_model'
dfx = clv_model.copy()

# Identify numeric features in the dataset
numeric_features = dfx.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Compute Pearson correlation for each numeric feature with the target 'Response'
pearson_correlations = {}
for feature in numeric_features:
    if feature != 'Response':
        # Ensure that both series do not contain NaN values and have at least 2 data points before calculation
        valid_data = dfx[[feature, 'Response']].dropna()
        if len(valid_data) >= 2:
            correlation, _ = pearsonr(valid_data[feature], valid_data['Response'])
            pearson_correlations[feature] = correlation

# Converting the results to a DataFrame
correlation_table = pd.DataFrame.from_dict(pearson_correlations, orient='index', columns=['Correlation'])

# Sort the table by correlation values
correlation_table.sort_values(by='Correlation', ascending=False, inplace=True)

# Display the DataFrame directly
correlation_table
```

#### Chi Square Test

```python
# Identify the categorical features in the dataset
categorical_columns = clv_model.select_dtypes(include=['object']).columns.tolist()
categorical_columns.remove('Customer')  # Exclude 'Customer' from the list

# Initialize a dictionary to store results of the Chi-square tests
chi2_test_results = {}

# Perform Chi-square tests on the categorical features
for column in categorical_columns:
    if column != 'Response':  # Ensure the target variable 'Response' is not included
        # Create a contingency table for the current feature and the target variable
        contingency_matrix = pd.crosstab(clv_model[column], clv_model['Response'])
        # Perform the Chi-square test
        chi2_stat, p_value, degrees_of_freedom, expected_values = chi2_contingency(contingency_matrix)
        # Store the results in the dictionary
        chi2_test_results[column] = {'Chi2 Statistic': chi2_stat, 'p-value': p_value}

# Convert the dictionary to a DataFrame for better visualization
chi2_results_df = pd.DataFrame(chi2_test_results).T

# Sort the DataFrame by the p-value column in ascending order
chi2_results_df.sort_values(by='p-value', inplace=True)

# Format the p-values to scientific notation with 3 decimal places
chi2_results_df['p-value'] = chi2_results_df['p-value'].map(lambda x: f'{x:.3e}')

# Apply conditional formatting to highlight p-values less than 0.05
styled_chi2_results_df = chi2_results_df.style.applymap(
    lambda val: 'color: white' if float(val) < 0.05 else 'color: blue', subset=pd.IndexSlice[:, ['p-value']]
)

# Display the styled DataFrame
styled_chi2_results_df
```

```python
# Identify categorical features in the dataset
categorical_features = clv_model.select_dtypes(include=['object']).columns.tolist()
categorical_features.remove('Customer')  # Remove 'Customer' from the list

# Define segmentation groups
segmentation_groups = {
    "Geographic Segmentation": ['State', 'Location Code'],
    "Demographic Segmentation": ['Gender', 'Marital Status', 'Education', 'EmploymentStatus'],
    "Policy Segmentation": ['Coverage', 'Policy Type', 'Policy', 'Renew Offer Type'],
    "Vehicle Segmentation": ['Vehicle Class', 'Vehicle Size'],
    "Performance Segmentation": ['Effective To Date', 'Complaint rate'],
    "Sales Segmentation" : ['Sales Channel']
}

# Perform cross-tabulation for each segmentation group
segmentation_results = {}

for segment, features in segmentation_groups.items():
    segment_results = {}
    for feature in features:
        if feature in categorical_features:  # Ensure the feature exists in the dataset
            cross_tab = pd.crosstab(clv_model[feature], clv_model['Response'], margins=False)
            
            # Filter to only include the column where Response is 1
            cross_tab = cross_tab[[1]] if 1 in cross_tab.columns else pd.DataFrame()

            if not cross_tab.empty:
                # Calculate percentages for Response = 1
                cross_tab_percentage = (cross_tab / cross_tab.sum(axis=0)) * 100
                cross_tab_combined = cross_tab.astype(str) + ' (' + cross_tab_percentage.round(2).astype(str) + '%)'

                segment_results[feature] = cross_tab_combined
    
    segmentation_results[segment] = segment_results

# Display the cross-tabulation results with percentages for each segmentation group
for segment, results in segmentation_results.items():
    print(f"\n{segment}:\n")
    for feature, cross_tab in results.items():
        print(f"Cross-tabulation for {feature}:\n{cross_tab}\n")
```
